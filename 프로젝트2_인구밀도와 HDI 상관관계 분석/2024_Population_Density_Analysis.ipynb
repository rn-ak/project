# 2024 인구밀도 분석

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
cols = ['population_2024','land_area', 'country', 'region', 'unMember' ]
```

```python
# 인구 밀도 및 토지 면적 비교_2024
world_populations = pd.read_csv('best-countries-to-live-in-2024.csv', usecols=cols)
world_populations
```

```python
# 인구밀도 계산
world_populations.insert(2, 'Population_density', (world_populations['population_2024'] / world_populations['land_area']).round(2))
world_populations
```

```python
# 나라인구밀도 정렬 
world_populations.sort_values('Population_density', ascending=False)
```

```python
# 나라 확인하기
world_populations['country'].unique()
```

# 상위, 하위 20개 국가 확인하기 

```python
# 상위 국가 
Densely_populated_country =world_populations.nlargest(n=20, columns=['Population_density'])
Densely_populated_country
```

```python
#시각화 
plt.figure(figsize=(21,6))
plt.bar(Densely_populated_country['country'], Densely_populated_country['Population_density'])
plt.xlabel('country')
plt.ylabel('Population_density')
plt.title('Densely_populated_country', fontsize=15)
plt.show()
```

```python
# 하위  국가 
Sparsely_populated_country = world_populations.nsmallest(n=20, columns=['Population_density'])
Sparsely_populated_country
```

```python
#시각화 
plt.figure(figsize=(21,6))
plt.bar(Sparsely_populated_country['country'], Sparsely_populated_country['Population_density'])
plt.xlabel('country')
plt.ylabel('Population_density')
plt.title('Sparsely_populated_country', fontsize=15)
plt.show()
```

# 세계지도 시각화 

```python
import geopandas as gpd
```

```python
# 세계지도 불러오기
# geometry : 멀티폴리곤으로 각 국가의 경계도 표시 
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world
```

```python
# 세계지도 데이터와 국가명 일치하게 변경
world_populations['country'] = world_populations['country'].replace({
    'United States': 'United States of America',
    'Bosnia and Herzegovina': 'Bosnia and Herz.',
    'Czech Republic': 'Czechia',
    'Republic of the Congo': 'Congo',
    'Ivory Coast' : 'Côte d\'Ivoire',
    'Eswatini' :'eSwatini'
})
```

```python
world_merge = world_populations[['country', 'Population_density']]
world_merge
```

```python
# 데이터 병합
world_df = pd.merge(world,world_merge, left_on='name', right_on='country', how='inner')
world_df
```

```python
world_df.plot(column='Population_density', colormap='inferno', legend=True, figsize=(15, 6))
plt.title('Population Density Map')
```

# 대륙별 인구밀도 분석하기 

```python
# 전세계 평균 인구밀도 구하기 
World_Population = 8.0e9 # 약 8억명
World_land_area = 1.48e8 # 약 1억 4800만 제곱킬로미터

Average_Population_Density = World_Population / World_land_area
Average_Population_Density
```

전세계 평균 인구밀도는 약 `54`명/제곱킬로미터

```python
# 평균보다 높은 인구밀도를 가진 국가
higher_Population_density = world_populations[world_populations['Population_density'] >= 54]
higher_Population_density
```

```python
higher_Population_density.info()
```

```python
# UN 회원국 여부
higher_Population_density['unMember'].value_counts()
```

```python
higher_Population_density[higher_Population_density['unMember'] == False]
```

```python
# 평균보다 낮은 인구밀도를 가진 국가
lower_Population_density = world_populations[world_populations['Population_density'] < 54]
lower_Population_density.head()
```

```python
lower_Population_density.info()
```

```python
# UN 회원국 여부
lower_Population_density['unMember'].value_counts()
```

### 전체 데이터 대륙별 국가수 확인

```python
# 전체 데이터 대륙수 
world_df_count = pd.DataFrame(world_populations['region'].value_counts())
world_df_count
```

```python
# 평균보다 인구밀도가 높은 대륙수 
higher_df = pd.DataFrame(higher_Population_density['region'].value_counts())
higher_df
```

```python
higher_df.plot(column='Population_density', colormap='inferno', legend=True, figsize=(10, 6))
plt.title('Continents with Above-Average Population Density')
```

```python
# 평균보다 인구밀도가 낮은 대륙수 
lower_df = pd.DataFrame(lower_Population_density['region'].value_counts())
lower_df
```

```python
lower_df.plot(column='Population_density', colormap='inferno', legend=True, figsize=(10, 6))
plt.title('Continents with Below-Average Population Density')
```
