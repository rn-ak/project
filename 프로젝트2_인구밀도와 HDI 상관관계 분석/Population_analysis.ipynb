### 전체 지역

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pop_region_growth=pop1.groupby(["region"], as_index=False).agg({"population_growthRate" : "mean", "Hdi2020" : "mean", "Hdi2021" : "mean"})
pop_region_growth
```

```python
pop_region_growth_oreders = pop_region_growth.sort_values(by="population_growthRate", ascending=False)["region"].to_list()
pop_region_growth_oreders
```

* 24년 기준으로 큰 기준인 region으로 나눠보니 평균 증감률이 0.7% ~ 2% 안팎으로 나타났습니다.
* 제일 높은 비중을 가진 나라는 Africa 였으며, Asia, North America, Oceania, South America 순으로 나타났습니다.
* Europe 지역은 전년 대비 감소되는 비율을 보였습니다.

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(pop_region_growth, x="region", y="population_growthRate", hue="region", order=pop_region_growth_oreders, palette="Set2")
for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.xlabel("Region")
plt.ylabel("Population_Growth")
plt.title("Average Population Growth by Region")
plt.show()
```

```python
pop_growth_melt = pop_region_growth.melt(id_vars=["region"], var_name="year", value_name="value")
pop_growth_melt
```

```python
pop_growth_melt[6:]
```

```python
pop_oceania
```

* 인구 평균 증감률이 높은 지역인 Africa는 제일 낮은 HDI 지수를 보였으며, Oceania 지역이 HDI 지수가 높은 것으로 나타났다.
* 하지만, oceania 데이터는 나라가 호주, 뉴질랜드인 2개 나라만 데이터가 확인이 되는 상태이기 때문에 추가 확인이 필요
* 이후에는 유럽이 제일 높았으며 근소한 차이로 남아메리카, 북아메리카, 아시아 지역으로 이어졌다.

```python
import numpy as np
hdi_region_range = np.arange(0, 1., 0.1)
hdi_region_range
```

```python
plt.figure(figsize=(20, 6))
ax = sns.barplot(data=pop_growth_melt[6:], x="region", y="value", hue="year", order=pop_region_growth_oreders, palette="Set2")
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.xlabel("Region")
plt.ylabel("HDI Point")
plt.yticks(hdi_region_range)
plt.legend(title="HDI index")
plt.title("HDI index by Region")
plt.show()
```
