```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_data = pd.read_csv('loss.csv')
loss_data
```

```python
# 날짜 형식 변환
loss_data['Date'] = pd.to_datetime(loss_data['Date'])
```

```python
# 오프라인 및 온라인 비용 시각화
plt.figure(figsize=(10, 6))
plt.plot(loss_data['Date'], loss_data['Offline_Cost'], label='Offline Cost', marker='o')
plt.plot(loss_data['Date'], loss_data['Online_Cost'], label='Online Cost', marker='o')
plt.xlabel('Date')
plt.ylabel('Cost')
plt.title('Offline vs. Online Cost over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Shipping Fee 막대 그래프 시각화
plt.figure(figsize=(10, 6))
plt.bar(loss_data['Date'], loss_data['Shipping_Fee'], color='b')
plt.xlabel('Date')
plt.ylabel('Shipping Fee')
plt.title('Shipping Fee over Time')
plt.grid(axis='y')
plt.show()
```

```python
# GST 막대 그래프 시각화
plt.figure(figsize=(10, 6))
plt.bar(loss_data['Date'], loss_data['GST'], color='g')
plt.xlabel('Date')
plt.ylabel('GST')
plt.title('GST over Time')
plt.grid(axis='y')
plt.show()
```
